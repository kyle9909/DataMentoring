```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
import seaborn as sns
import matplotlib.pyplot as plt
import sys

#한글 인코딩
from matplotlib import rc
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
from datetime import datetime
```

```python
gu_num = pd.read_csv('구별충전기개수.csv')
gu_num = gu_num.sort_values(by='한글주소')
gu_num.reset_index(inplace=True)
gu_num.head()
gu_num.head()
```

```python
all = pd.read_csv('총정리df.csv')
all = all.sort_values(by='자치구')
all.reset_index(inplace=True)
all.head()
```

```python
env_to_all = pd.concat([all, gu_num[['충전기 개수']]], axis=1)
env_to_all.head()
```

```python
for i in range(len(env_to_all)):
    env_to_all.loc[i,'전체/한황'] = env_to_all.loc[i,'충전기개수'] / env_to_all.loc[i,'충전기 개수']
    
env_to_all.head()
```

```python
env_to_all = pd.concat([gu_num, env_to_all[['전체/한황']]], axis=1)
env_to_all.head()
```
